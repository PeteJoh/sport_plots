```python
import pandas as pd
import numpy as np
from PIL import Image
import urllib
import math
import matplotlib.colors as mcolors
from matplotlib import cm
import matplotlib.ticker as ticker
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from highlight_text import fig_text
import os
```

```python
font_path = "assets/fonts"
for x in os.listdir(font_path):
    for y in os.listdir(f"{font_path}/{x}"):
        if y.split(".")[-1] == "ttf":
            fm.fontManager.addfont(f"{font_path}/{x}/{y}")
            try:
                fm.FontProperties(weight=y.split("-")[-1].split(".")[0].lower(), fname=y.split("-")[0])
            except Exception:
                continue

#plt.style.use("../assets/stylesheets/soc_base.mplstyle")
```

```python
df=pd.read_html("https://fbref.com/en/comps/9/Premier-League-Stats")[0]
```

```python
df.set_index("Rk", inplace=True)
```

```python
df=df[["Squad","GF","xG"]]
```

```python
df.dtypes
```

```python
df["difference"]=df["GF"]-df["xG"]
df.sort_values("difference", inplace=True,ascending=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
def ax_logo(team_id, ax):
    '''
    Plots the logo of the team at a specific axes.
    Args:
        team_id (int): the id of the team according to Fotmob. You can find it in the url of the team page.
        ax (object): the matplotlib axes where we'll draw the image.
    '''
    fotmob_url = 'https://images.fotmob.com/image_resources/logo/teamlogo/'
    url=f'{fotmob_url}{team_id}'
    #print(url)
    club_icon = Image.open(urllib.request.urlopen(url))#convert("LA")
    ax.imshow(club_icon)
    ax.axis('off')
    return ax
```

```python
background_color="#e3e0d8" #"#cfcdc6"
fig = plt.figure(figsize=(7,10), dpi=75,facecolor =background_color  )#"#dbd6d0")
ax = plt.subplot(facecolor = background_color)#"#dbd6d0" ,)




max_ = df["difference"].abs().max()*1.05
ax.set_xlim(-14.5, max_+1.8)
ax.set_ylim(-1, df.shape[0])
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.rcParams['hatch.linewidth'] = 0.75

ax.yaxis.set_ticks([])
ax.grid(ls='--', lw=1, color='lightgrey', axis='x')
norm = mcolors.Normalize(vmin=df['difference'].min(),vmax=df['difference'].max())
cmap = plt.get_cmap('SOC')

ax.barh(df.index,df['difference'],zorder=3,color = cmap(norm(df['difference'])),ec='#efe9e6',hatch='//////')

#middle vertical line
ax.plot(
    [0,0],
    [ax.get_ylim()[0], ax.get_ylim()[1]-0.5],
    color='black',
    lw=.75,
    zorder=3,ls=":"
)

"""#middle  horizontal line
ax.plot(
    [ax.get_xlim()[0],ax.get_xlim()[1]],
    [9.5,9.5],
    color='black',
    lw=.75,
    zorder=3,ls=":"
)"""
#top horizontal line
"""ax.plot(
    [ax.get_xlim()[0],ax.get_xlim()[1]],
    [df.shape[0]-0.5,df.shape[0]-0.5],
    color='black',
    lw=.75,
    zorder=3
)"""

#add club logo

DC_to_FC = ax.transData.transform
FC_to_NFC = fig.transFigure.inverted().transform
DC_to_NFC = lambda x: FC_to_NFC(DC_to_FC(x))


#annotate
for index, x in enumerate(df['difference']):
    if x < 0:
        sign_offset = -1
        sign_text = ''
        x_pos=0.5
    else:
        sign_offset = 1
        sign_text = '+'
        x_pos=-1.8
    text_ = ax.annotate(
        xy=(x, index),
        xytext=(sign_offset*18,0),
        text=f'{sign_text}{x:.1f}',
        ha='center',
        va='center',
        textcoords='offset points'
    )
    ax_coords = DC_to_NFC([x_pos, index - 0.5])
    logo_ax = fig.add_axes([ax_coords[0], ax_coords[1], 0.035, 0.035], anchor = "C")
    ax_logo(club_logos[index],logo_ax)


ax.xaxis.set_major_locator(ticker.MultipleLocator(2))



#heading
fig_text(
    x = 0.26, y = 0.94,
    s = "Clubs xG Performance",
    color = "black",
    size = 16, weight="bold",font = 'DM Sans',
    annotationbbox_kw={"xycoords": "figure fraction"}
)
#subheading
fig_text(
    x = 0.26, y = 0.91,
    s = "2022/23 Season | EPL stats alone |  by @saheedniyi",
    color = "#3d3d3d",
    size = 10, weight="bold",font = 'Karla',
    annotationbbox_kw={"xycoords": "figure fraction"}
)


#image url
epl_logo_url = "https://images.fotmob.com/image_resources/logo/leaguelogo/47.png"
#Get and Open the league logo
league_logo = Image.open(urllib.request.urlopen(epl_logo_url))
#create new axes at top right cornier [left_position,bottom_position,width,hight]
logo_ax = fig.add_axes([0.15, 0.87,0.1, 0.1])
#paste logo on the axes
logo_ax.imshow(league_logo)
#switch off  the axis off our AXES
logo_ax.axis("off")

# -- Figure title and arrows --------------------------------
position_negative = df[df['difference'] < 0].index.max()
position_x_negative = math.floor(-max_*(.75))
position_x_positive = math.ceil(max_*(.85))
ax.annotate(
    xy=(position_x_negative,position_negative + .5),
    xytext=(position_x_negative,ax.get_ylim()[1]),
    text='',
    arrowprops=dict(arrowstyle='<-',color='gray')
)
ax.annotate(
    xy=(position_x_positive,position_negative),
    xytext=(position_x_positive,ax.get_ylim()[0] + .2),
    text='',
    arrowprops=dict(arrowstyle='<-',color='gray')
)
mid_point_positive = (position_negative + ax.get_ylim()[1])/2
mid_point_negative = (position_negative + ax.get_ylim()[0])/2
ax.annotate(
    xy=(position_x_negative,mid_point_positive),
    text='xG over performers',
    rotation=90,
    xytext=(-10,0),
    textcoords='offset points',
    ha='center',
    va='center',
    size=12
)
ax.annotate(
    xy=(position_x_positive,mid_point_negative),
    text='xG under performers',
    rotation=-90,
    xytext=(10,0),
    textcoords='offset points',
    ha='center',
    va='center',
    size=12
)
```
